```python
#Cloud
graph = Graph(scheme="bolt", host="54.90.165.10", port="7687", user="neo4j", password="clicks-haul-qualifier")
```

```python
query = '''
MATCH (stock)-[r:TRADING_HOUR]->(stockQuote) 
RETURN stock.name,
stockQuote.timestamp_utc,
stockQuote.open,
stockQuote.high,
stockQuote.low,
stockQuote.close,
stockQuote.volume
'''
result = graph.query(query)
stockQuotesDf = pd.DataFrame(result)
stockQuotesDf.rename(columns={0:'ticker', 1:'timestamp_utc', 2:'open', 3:'high', 
                        4:'low', 5:'close', 6:'volume'}, inplace=True)
```
